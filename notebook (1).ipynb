```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Filter data for movies released in the 1990s
movies_df=netflix_df[["title","release_year"]]
movies_fil=movies_df[(movies_df["release_year"]>=1990) & (movies_df["release_year"]<=1999)]
print(movies_fil)
```

```python
#Find the most frequent movie duration
duration_df=netflix_df[["duration"]]
plt.figure(figsize=(10, 6))
plt.hist(duration_df, bins=30, edgecolor='k', color='limegreen')
plt.title("Distribution of Movie Durations", fontsize=16)
plt.xlabel("Duration (minutes)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
duration=netflix_df["duration"].value_counts().idxmax()
print("The most frequent duration is: "+ str(duration))
```

```python
#Count the number of short action movies from the 1990s
movies_df = netflix_df[["title", "genre","duration","release_year"]]
movies_act=movies_df[movies_df["genre"]=="Action"]
movies_act90 =movies_act[(movies_act["release_year"]>=1990) & (movies_act["release_year"]<=1999)]
short_movie_count = 0
for label, row in movies_act90.iterrows():
    if row["duration"] < 90:
        short_movie_count+= 1

print("Count of short action movies: " +str(short_movie_count))
```
